```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
plt.plot(x,y);
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('First plot')
plt.show()
```

```python
fig = plt.figure(figsize=(3,3))
axes = fig.add_axes([0,0,0.5,0.5])
axes.plot(x,y)
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
fig = plt.figure(figsize=(2,2))
#large plot
axes1 = fig.add_axes([0,0,2,2])
axes1.plot(a,b)
plt.title('power of A')
#small plot
axes2 = fig.add_axes([0.5,1,0.5,0.5])
axes2.plot(a,b)
plt.xlim(1,2)
plt.ylim(0,50)
plt.title('Zoomed In')
```

```python
fig = plt.figure(figsize=(5,4),dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
fig.savefig('secondplot.png',bbox_inches='tight')
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_title('Linear Line')
axes[0][1].plot(x,y)
axes[1][1].plot(a,b)
axes[1][1].set_title('Power of A')
axes[1][0].plot(a,b)
fig.suptitle('Subplots')
plt.tight_layout()
fig.savefig('Firstsubplots',bbox_inches='tight')
```

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='X vs X^2')
ax.legend(loc=(1.1,0.8))
```

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='green',label='X vs X')
ax.plot(x,x+1,color='red',label='X vs X+1',lw=3)
ax.grid(color='blue',alpha=0.5,ls='dashed',lw=0.5)
ax.legend()
```
